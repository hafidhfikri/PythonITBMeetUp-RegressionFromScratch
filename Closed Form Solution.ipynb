## Python ITB - Regression From Scratch

```python
#Import Everything
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-white')
```

```python
#Load Data
house = datasets.load_boston()
X = house.data[:,5]
Y = house.target
```

```python
#Visualize Data
plt.title('Boston House Pricing')
plt.xlabel('Number of rooms')
plt.ylabel('House price')
plt.scatter(X,Y)
```

```python
#b1 function
n = X.shape[0]
b1 = (np.sum(X*Y) - (1/n*np.sum(X)*np.sum(Y))) /((np.sum(X*X) - (1/n*np.sum(X)*np.sum(X)))) 
```

```python
#b0 function
b0 = np.mean(Y) - (b1*np.mean(X))
```

```python
#calculate y estimate
y_estimate = b0 + (b1*X)
```

```python
#visualize estimateion
plt.plot(X,y_estimate)
plt.title('Boston House Pricing')
plt.xlabel('Number of rooms')
plt.ylabel('House price')
plt.scatter(X,Y)
```

```python
RSS = np.sum((Y-y_estimate)**2)/n
print(RSS)
```
